```python
#Create new CSV File and write to it
def writeToCSV(path, header, data):
    with open(path, 'w', encoding='UTF8', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(header)
        writer.writerow(data)
        
#Append existing CSV File
def appendCSV(path, data):
    with open(path, 'a', encoding='UTF8', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(data)

#Get and save listing and profile data
def getContent(driver, url):
    listingHeaders = ['Description','Location','ImageCount','ImageQuality']
    profileHeaders = ['ID','URL']
    listingInformation = []
    profileInformation = []
    
    #Open marketplace page
    driver.get(url)
    
    #Get Listing description
    description = driver.find_element_by_xpath("(//span[@class='x193iq5w xeuugli x13faqbe x1vvkbs x1xmvt09 x1lliihq x1s928wv xhkezso x1gmr53x x1cpjm7i x1fgarty x1943h6x xudqn12 x3x7a5m x6prxxf xvq8zen xo1l8bm xzsf02u'])")

    #Get Listing location
    location = driver.find_element_by_xpath("(//span[@class='x193iq5w xeuugli x13faqbe x1vvkbs x1xmvt09 x1lliihq x1s928wv xhkezso x1gmr53x x1cpjm7i x1fgarty x1943h6x x4zkp8e x676frb x1nxh6w3 x1sibtaa xo1l8bm xi81zsa'])")
    try:
        location = location.text.split(": ",1)[1]
    except:
        location = location.text.split(" in ",1)[1]
        
    #Get images data
    images = driver.find_elements_by_xpath("(//div[@class='x6s0dn4 x78zum5 x193iq5w x1y1aw1k xwib8y2 xu6gjpd x11xpdln x1r7x56h xuxw1ft xc9qbxq']//img)")

    if len(images) == 0:
        images = driver.find_elements_by_xpath("(//div[@class='x6s0dn4 x78zum5 x1iyjqo2 xl56j7k x6ikm8r x10wlt62 xh8yej3 x1ja2u2z']//img)")

    image_size = []
        
    for image in images:
        h = image.get_attribute('naturalHeight')
        w = image.get_attribute('naturalWidth')
        image_size.append(str(w)+"x"+str(h))
        
    #Save Listing Information to Array
    listingInformation.append(description.text)
    listingInformation.append(location)
    listingInformation.append(len(images))
    listingInformation.append(image_size)
    
    #Click User Profile
    button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "div[class='x1i10hfl xjbqb8w x6umtig x1b1mbwd xaqea5y xav7gou x9f619 x1ypdohk xt0psk2 xe8uvvx xdj266r x11i5rnm xat24cr x1mh8g0r xexx8yu x4uap5 x18d9i69 xkhd6sd x16tdsg8 x1hl2dhg xggy1nq x1a2a7pz xt0b8zv xzsf02u x1s688f']"))).click()
    
    time.sleep(2)

    #Get User Profile URL
    profile = driver.find_element_by_xpath("//div[@class='__fb-light-mode x1n2onr6 xzkaem6']//a[@class='x1i10hfl xjbqb8w x6umtig x1b1mbwd xaqea5y xav7gou x1ypdohk xe8uvvx xdj266r x11i5rnm xat24cr x1mh8g0r xexx8yu x4uap5 x18d9i69 xkhd6sd x16tdsg8 x1hl2dhg xggy1nq x1o1ewxj x3x9cwd x1e5q0jg x13rtm0m x87ps6o x1lku1pv x1a2a7pz x9f619 x3nfvp2 xdt5ytf xl56j7k x1n2onr6 xh8yej3']")
    profileUrl = profile.get_attribute('href')

    #Destructure url to get Profile ID
    ID = profileUrl.rsplit("/",1)[1]
    
    #Save Profile ID and url to Array
    profileInformation.append(ID)
    profileInformation.append(profileUrl)
    
    #Paths to CSV Files
    listingPath = '../../zorko/Desktop/Uni/Year3-Project/CSV_Files/Listings.csv'
    profilePath = '../../zorko/Desktop/Uni/Year3-Project/CSV_Files/Profiles.csv'

    #Check if CSV Files exist and save Arrays to CSV Files
    write = True

    if os.path.exists(listingPath):
        write = False

    if write:
        writeToCSV(listingPath,listingHeaders,listingInformation)
    else:
        appendCSV(listingPath,listingInformation)

    write = True

    if os.path.exists(profilePath):
        write = False

    if write:
        writeToCSV(profilePath,profileHeaders,profileInformation)
    else:
        appendCSV(profilePath,profileInformation)
```
