```python
xs = ""
fb_dtsg = ""
queries = ['; c_user=','; xs=','&fb_dtsg=']
data = {"user_id":"","xs":"","fb_dtsg":""}
test = "test"

for log in log2:
    log = log["message"]
    if len(queries) == 0:
        break
    for query in queries:
        idx = log.find(query)
        if idx != -1:
            shortLog = log[idx+len(query):]
            if query == '&fb_dtsg=':
                data["fb_dtsg"] = shortLog.split("&")[0]
                queries.remove(query)
            else:
                newData = shortLog.split(";")[0]
                if query == "; xs=":
                    data["xs"] = newData
                else:
                    data["user_id"] = newData
                queries.remove(query)
```

```python
template = {"user_id": "",
            "cookies": "",
            "fb_dtsg": ""}

cookies = f'c_user={data["user_id"]}; xs={data["xs"]}'

for key in template.keys():
    if key == "cookies":
        template[key] = cookies
        continue
    template[key] = data[key]
```
